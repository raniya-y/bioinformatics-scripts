```python
# task 1

from Bio.Seq import Seq 
dna = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")
 rna = dna.transcribe() # DNA → RNA (T replaced with U)
 print("RNA:", rna) 
protein = rna.translate() # RNA → Protein (codons → amino acids) print("Protein:", protein)
 print("Number of amino acids:", len(protein)) # Count amino acids 
```
