```python
# Q1: DNA Sequence Storage Using Lists

# Store the sequences in a list
sequences = ["ATGCGT", "TTAACG", "CGTACG", "GGCATC"]

# Print the total number of sequences
print("Total number of sequences:", len(sequences))

# Print the third sequence in the list
print("Third sequence in the list:", sequences[2])
```

```python
# Q2: Codon Lookup Using Tuples


codon_table = [("AUG", "Methionine"), ("UUU", "Phenylalanine"), ("UUC", "Phenylalanine")]
codon_set = set(codon_table)

print("Codon set:", codon_set)

print("Number of codon pairs:", len(codon_set))
```

```python
# Q3: DNA Sequence Lengths Using Dictionary

# Dictionary of sequences
sequences = {"A": "ATGCGT", "B": "TTAACGGA", "C": "CGTACGTTAA"}

print("Length of sequence B:", len(sequences["B"]))
print("All sequence names:", sequences.keys())
print("All DNA sequences:", sequences.values())
```

```python
# Q4: Mutation Tracking Using Sets

# Set of mutation identifiers
mutations = {"MUT1", "MUT2", "MUT3", "MUT4", "MUT5"}

print("Total number of mutations:", len(mutations))
print("Is MUT3 in the set?", "MUT3" in mutations)
print("All mutations:", mutations)
```

```python
# Q5: FASTA Header Mapping Using Dictionary

# Dictionary mapping FASTA headers to sequences
fasta = {">seq1": "ATGCGTAC", ">seq2": "TTAACGGA", ">seq3": "GGCATCGT"}

print("Sequence for >seq2:", fasta[">seq2"])

print("All headers:", fasta.keys())

print("All sequences:", fasta.values())
```
